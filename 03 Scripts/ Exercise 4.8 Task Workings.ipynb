## 4. Create a loyalty flag for existing customers 

```python
# Use groupby and transform() functions with np.max argument to determine maximum order number for each user
df_ords_prods_merge['max_order']=df_ords_prods_merge.groupby(['user_id'])['order_sequence_number'].transform(np.max)
```

```python
# view updated data frame 
df_ords_prods_merge.head(15)
```

```python
# Use loc function to make loyalty flag column with given criteria
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check frequency of values in new loyalty flag column
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# View updated data frame
df_ords_prods_merge.head()
```

## 5.Look at spending habits of loyalty groups 

### Question 5

##### The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
# Use groupby and agg() to determine basic statisitcs of spending habits by loyalty group
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

##### The average price of a product purchased by a new customer is about $3 more than that of a loyal customer.

```python
# Use transform() to determine the mean price of products purchased by each user
df_ords_prods_merge['mean_product_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# View relevant columns 
df_ords_prods_merge[['user_id', 'mean_product_price']].head(20)
```

```python
# Create spending_flag column with given criteria using loc() function
df_ords_prods_merge.loc[df_ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#Check frequency of new spending_flag column
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# View relevant columns of updated data frame
df_ords_prods_merge[['user_id', 'mean_product_price', 'spending_flag']].head(20)
```

```python
# Use groupby and transform to determine median number of days between orders for each user
df_ords_prods_merge['median_days_between_orders'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# View relevant columns of updated dataframe
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head(20)
```

```python
# Create a freaquency flag for the customer using the loc function.
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_between_orders'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_between_orders'] <= 20) & (df_ords_prods_merge['median_days_between_orders'] > 10), 'order_frequency'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_between_orders'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# Check value counts of new order_frequency column
df_ords_prods_merge['order_frequency'].value_counts(dropna = False)
```

```python
# View relevant columns of updated data frame
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders', 'order_frequency']].head(25)
```
